```python
import scraper as s
import pandas as pd
import numpy as np
import datetime
```

```python
m = s.stockMongo()
apple = m.get_options("AAPL")
```

```python
apple.tail()
```

```python
#apple = apple.where(apple['last_trade']==apple.index.date).dropna()
apple = apple.drop(['volume', 'iv', 'in-money', 'bid', 'ask','last_trade', 'contract'], axis=1)
```

```python
apple = apple.reset_index()
apple = apple.set_index(['type', 'strike', 'strike-date'])
apple['date'] = pd.to_datetime(apple['date']).dt.strftime('%m-%d-%Y')
apple['date'] = pd.to_datetime(apple['date'])
#apple.loc[('AAPL210917C00190000', 'Call')]
aC350 = apple.loc[('Call', 380,'2022-09-16')]
aC350 = aC350.reset_index()
aC350 = aC350.drop(['type', 'strike-date'], axis=1)
aC350.rename(columns = {'last':'priceC350', 'strike': 'strikeC350'}, inplace = True)
aC360 = apple.loc[('Call', 390,'2022-09-16')]
aC360 = aC360.reset_index()
aC360 = aC360.drop(['type', 'strike-date'], axis=1)
aC360.rename(columns = {'last':'priceC360', 'strike': 'strikeC360'}, inplace = True)
aP350 = apple.loc[('Put', 350,'2022-09-16')]
aP350 = aP350.reset_index()
aP350 = aP350.drop(['type', 'strike-date'], axis=1)
aP350.rename(columns = {'last':'priceP350', 'strike': 'strikeP350'}, inplace = True)
aP360 = apple.loc[('Put', 360,'2022-09-16')]
aP360 = aP360.reset_index()
aP360 = aP360.drop(['type', 'strike-date'], axis=1)
aP360.rename(columns = {'last':'priceP360', 'strike': 'strikeP360'}, inplace = True)

aP360.head()
```

```python
apl = m.get_finnhub_prices('AAPL', min(aC360['date']), max(aC360['date']))
apl['date'] = apl.index
apl = apl.reset_index()
apl = apl.drop(['h', 'l', 'o', 'v','t'], axis=1)
a = pd.merge(aC360, apl, on='date')
a = pd.merge(a, aC350, on='date')
a = pd.merge(a, aP350, on='date')
a = pd.merge(a, aP360, on='date')
a.head()
```

```python
def iron_condor(x):
    s = -x['priceC360']+x['priceC350']+x['priceP360']-x['priceP350']
    if x['c'] >= x['strikeC350']:
        s = s - x['c'] + x['strikeC350']
    if x['c'] >= x['strikeC360']:
        s = s + x['c'] - x['strikeC360']
    if x['c'] <= x['strikeP360']:
        s = s - x['c'] + x['strikeP360']
    if x['c'] <= x['strikeP350']:
        s = s + x['c'] - x['strikeP350']
    #/(x['strikeP360']-x['strikeP350'])
    return s
```

```python
a['daily_return'] = a.apply(iron_condor, axis=1)
a['value'] = -a['priceC360']+a['priceC350']+a['priceP360']-a['priceP350']
a['position'] = - a['c'] + a['strikeC350']
```
