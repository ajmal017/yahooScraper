```python
apl = m.get_stock_data('AAPL')
```

```python
def split(apple):
    if apple['date'] < datetime.fromisoformat('2020-08-31'):
        return apple['c'] * 4
    else:
        return apple['c']
```

```python
apple2 = apple.reset_index()
apple2['date'] = apple2['date'].dt.normalize()
apple2 = pd.merge(apple2, apl[['c', 'date']], on='date')
#apple2 = apple2.set_index('date')
apple2['c'] = apple2.apply(split, axis=1)
apple2.head()
```

```python
apple_calls = apple2.where(apple2['type']=="Call").dropna()
apple_puts = apple2.where(apple2['type']=="Put").dropna()
apple_calls = apple_calls.where(apple_calls[])
apple_calls.head()
```
